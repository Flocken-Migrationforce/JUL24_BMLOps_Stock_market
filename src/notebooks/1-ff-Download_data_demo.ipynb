Get demo data

```python
import requests
import numpy as np
import pandas as pd

# replace the "demo" apikey below with your own key from https://www.alphavantage.co/support/#api-key
url = 'https://www.alphavantage.co/query?function=HISTORICAL_OPTIONS&symbol=IBM&apikey=demo'
r = requests.get(url)
data = r.json()

print(data)
```

Export to file :    

```python
import csv, os

# Define the path to the 'data' folder
data_folder = '../data'  # Move one level up to the parent directory and then into 'data'
file_path = os.path.join(data_folder, '1.1-demo.csv')

# Write the data to the CSV file
with open(file_path, mode='w', newline='') as file:
    writer = csv.writer(file)

    # Write the header (keys of the JSON data)
    writer.writerow(data.keys())

    # Write the values (data corresponding to the keys)
    writer.writerow(data.values())

print(f"Data has been written to: {file_path}")
```
